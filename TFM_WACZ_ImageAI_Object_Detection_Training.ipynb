CROP IMAGES
En a siguinte sección se toman las coordenadas de un bounding box que contenga un leucocito o WBC, y se recorta la imagen para luego identificar el tipo de leucocito con el CLASIFICADOR.

```
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Importing Image class from PIL module 
from PIL import Image 
  
# Opens a image in RGB mode 
im = Image.open(r"/content/drive/My Drive/Colab Notebooks/Blood-Cell-Count-Dataset/image2-detected.jpg") 
# Shows the image in image viewer 
plt.imshow(im)
plt.axis('Off')
plt.show()  
# Size of the image in pixels (size of orginal image) 
# (This is not mandatory) 
width, height = im.size
print('width:', width,'height: ',height)
```

```
#Funcion que ajusta la imagen recortada dadas las coordenadas xmin,ymin,xmax,ymax, a un tamaño de 320x240
#Programada por Wilson Castro Z.
def imrecortada(xmin,ymin,xmax,ymax,width,height):
  ancho=xmax-xmin
  alto=ymin-ymax
  while(ancho<320):
    if(xmin-1>=0):
      xmin=xmin-1
    xmaximo=width-xmin
    if(xmax+1<=xmaximo):
      xmax=xmax+1
    ancho=xmax-xmin
  while(alto<240):
    if(ymin-1>=0):
      ymin=ymin-1
    ymaximo=height-ymin
    if(ymax+1<ymaximo):
       ymax=ymax+1
    alto=ymax-ymin
  return xmin,ymin,xmax,ymax
```

```
#Code 1 (Continuacion)
#Como el codigo original no muestra las imagenes, usar matplotlib 
#%matplotlib inline

#import matplotlib.pyplot as plt
#import matplotlib.image as mpimg
# Importing Image class from PIL module 
#from PIL import Image 
  
# Opens a image in RGB mode 
# im = Image.open(r"/content/drive/My Drive/Colab Notebooks/geeks8-300x66.png") 
  
# Size of the image in pixels (size of orginal image) 
# (This is not mandatory) 
#width, height = im.size 
  
# Setting the points for cropped image 
#197, 123, 433, 333
width=640
height=480
left = 197
top = 123
right = 433
bottom = 333
#Funcion que ajusta el area de la imagen recortada a aprox 320*240
left,top,right,bottom=imrecortada(left,top,right,bottom,width,height)

 
# Cropped image of above dimension 
# (It will not change orginal image) 
im1 = im.crop((left, top, right, bottom)) 
  
# Shows the image
plt.imshow(im1)
plt.axis('Off')
plt.show()
#plt.savefig('/content/drive/My Drive/Colab Notebooks/im1.png')
```

```
#Reducir la imagen a 120x160 (la mitad para correrla al clasificador)
# Downsize the image with an ANTIALIAS filter (gives the highest quality)
 im1 = im1.resize((160,120),Image.ANTIALIAS)
 #Show the new image
 plt.imshow(im1)
plt.axis('Off')
plt.show()
```

CARGAR EL CLASIFICADOR

```
#se carga el arhivo de la ruta :'/content/drive/My Drive/Colab Notebooks/saved_model.pb'

new_model = tf.keras.models.load_model('/content/drive/My Drive/Colab Notebooks/')

# Check its architecture
new_model.summary()
```

```
#Cargar la imagen recortada para clasificarla segun: 
# - eosinophil: [1.0.0.0.]
# - lymphocyte: [0.1.0.0.]
# - monocyte:   [0.0.1.0.]
# - neutrophil: [0.0.0.1.]
import numpy as np
from google.colab import files
from keras.preprocessing import image

x = image.img_to_array(im1)
x = np.expand_dims(x, axis=0)

images = np.vstack([x])
classes = new_model.predict(images, batch_size=10)

print(classes)
x=new_model.predict_classes(images, batch_size=10)
print(x)
if(x==3):
  print('neutrophil')
elif(x==2):
  print('monocyte')
elif(x==1):
  print('lymphocite')
else:
  print('eosinophil')
```
